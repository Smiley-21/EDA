Combining Country Code and Zomato Dataset


```python
final_df=pd.merge(df, df_country, on='Country Code', how='left')
df.head()
```

```python
final_df.head()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
### Pie Chart
plt.pie(country_val,labels=country_name)
```

Above Pie Char are Complicated

```python
plt.pie(country_val[:3], labels=country_name[:3], autopct='%1.2f%%')
```

Most of the Zomato transaction are taking place in India after that UK and then US

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text'])
final_df
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

#Observation
1. When rating is above 4.5  -->Excellent
2. When rating is between 4.0 to 4.4 -->Very Good
3. When rating is between 3.5 to 3.9 -->Good
4. When rating is between 2.5 to 3.4 -->Average

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

Try to color bar according to color of ratings provided in the data

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color', palette=['blue', 'red','orange','yellow','lightgreen','darkgreen'])
```

```python
## Count Plot
sns.countplot(x='Rating color',data=ratings,palette=['blue', 'red','orange','yellow','lightgreen','darkgreen'])
```
